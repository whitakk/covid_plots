```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from datetime import datetime

import pickle
```

```python
# daily vaccination data from Our World In Data
vax = pd.read_csv('owid/us_state_vaccinations.csv')
vax.tail()
```

```python
vax = vax[['date','location','people_fully_vaccinated_per_hundred']]
vax.columns=['date','state','vaccinated_100']
```

```python
vax['vaccinated_100'] = vax.groupby('state')['vaccinated_100'].fillna(method='ffill')
```

```python
vax['state'].unique()
```

```python
us = vax[vax['state']=='United States']
ny = vax[vax['state']=='New York State']
nj = vax[vax['state']=='New Jersey']
```

```python
import pickle
pickle.dump(us, open('owid/us_vax.p', 'wb'))
pickle.dump(ny, open('owid/ny_vax.p', 'wb'))
pickle.dump(nj, open('owid/nj_vax.p', 'wb'))
```
