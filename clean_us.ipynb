```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from datetime import datetime

import pickle
```

```python
# national and state-level cases from NYT repo
us = pd.read_csv('covid-19-data/us.csv')
us.head()

us['cases_1'] = us['cases'].shift(1).fillna(0)
us['cases_delta'] = us['cases'] - us['cases_1']
us['cases_7'] = us['cases_delta'].rolling(7).mean()
us['cases_100k'] = us['cases_7']*100/328000
us.tail()
```

```python
ny = pd.read_csv('covid-19-data/us-states.csv')
ny = ny[ny['state']=='New York']

ny['cases_1'] = ny['cases'].shift(1).fillna(0)
ny['cases_delta'] = ny['cases'] - ny['cases_1']
ny['cases_7'] = ny['cases_delta'].rolling(7).mean()
ny['cases_100k'] = ny['cases_7']*100/19450
ny.tail()
```

```python
nj = pd.read_csv('covid-19-data/us-states.csv')
nj = nj[nj['state']=='New Jersey']

nj['cases_1'] = nj['cases'].shift(1).fillna(0)
nj['cases_delta'] = nj['cases'] - nj['cases_1']

# **ad-hoc error correction **
fill = nj[nj['date']=='2021-01-03']['cases_delta'].values[0]
nj['cases_delta'] = np.where(nj['date']=='2021-01-04', fill, nj['cases_delta'])
# ****

nj['cases_7'] = nj['cases_delta'].rolling(7).mean()
nj['cases_100k'] = nj['cases_7']*100/8882
nj.tail()
```

```python
nyc = pd.read_csv('covid-19-data/us-counties.csv')
nyc = nyc[nyc['county']=='New York City']

nyc['cases_1'] = nyc['cases'].shift(1).fillna(0)
nyc['cases_delta'] = nyc['cases'] - nyc['cases_1']
nyc['cases_7'] = nyc['cases_delta'].rolling(7).mean()
nyc['cases_100k'] = nyc['cases_7']*100/8419
nyc.tail()
```

```python
import pickle
pickle.dump(us, open('covid-19-data/us.p', 'wb'))
pickle.dump(ny, open('covid-19-data/ny.p', 'wb'))
pickle.dump(nj, open('covid-19-data/nj.p', 'wb'))
pickle.dump(nyc, open('covid-19-data/nyc.p', 'wb'))
```
